```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# création d'un un DataFrame à partir de nos données
data_df = pd.DataFrame(data, columns=['durée']) 

sns.countplot(x='durée', data=data_df)
plt.title('Décompte des valeurs de durée')
plt.xlabel('Durée')
plt.ylabel('Nombre d\'occurrences')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['store_and_fwd_flag'] = le.fit_transform(data['store_and_fwd_flag'])
```

```python
x=data.drop(['total_amount','fare_amount','extra','mta_tax','tip_amount','tolls_amount','improvement_surcharge','tpep_dropoff_datetime','tpep_pickup_datetime'],axis=1)
```

```python
corr_df = x.corr(method='pearson')

plt.figure(figsize=(12, 10))
sns.heatmap(corr_df, annot=True)
plt.show()
```

```python
x
```

```python
y=data['total_amount']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test= train_test_split(x,y,test_size=0.2)
```

```python
#régression linéaire
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from matplotlib.pyplot import *
%matplotlib inline
import numpy as np

lr = LinearRegression() # Instanciation

lr.fit(x_train,y_train) # Entraînement du modèle
Yp = lr.predict(x_test) # Test

```

```python
# A vous

print("Le socre de détermination est", lr.score(x_test, y_test))  #R squared est
# une mesure de la qualité de la prédiction d'une régression linéaire 
from sklearn.metrics import r2_score
print("R squared ", r2_score(y_test, Yp))
```

```python
import matplotlib.pyplot as plt

# Calcul des résidus
residuals = y_test - Yp

# Affichage du graphique de dispersion
plt.scatter(Yp, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Valeurs prédites')
plt.ylabel('Résidus')
plt.title('Graphique de dispersion des résidus')
plt.show()
```

nous constatons un R2 bon et très rapide temps d'éxécution

```python
from tensorflow import keras
model_network = keras.Sequential()
model_network.add(keras.layers.Dense(units = 64, activation = 'relu', input_dim=x_train.shape[1]))
model_network.add(keras.layers.Dense(units = 32, activation = 'relu'))
model_network.add(keras.layers.Dense(units = 16, activation = 'relu'))
model_network.add(keras.layers.Dense(units = 8, activation = 'relu'))
model_network.add(keras.layers.Dense(units = 1, activation = 'linear'))
model_network.compile(loss='mse', optimizer="adam")
```

```python
# Display the model
model_network.summary()
  
# Training
model_network.fit(x_train, y_train, epochs=10, verbose=1, batch_size=64)
```

```python
  
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred = model_network.predict(x_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(rmse, mae , r2)
```
